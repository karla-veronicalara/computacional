```python
import numpy as np
import folium
import matplotlib.pyplot as plt
from IPython.display import display, IFrame
import os
```

```python
def mi_lista_csv(varios_archivos):
    filtro = []
    for nombre in varios_archivos:
        if nombre[-4:] == '.csv': #Compara los ùltimos 4 caracteres de cada archivo
            filtro.append(nombre)
    return filtro

os.system('ls')
```

```python
lista = os.listdir('./ArchivosCSV/')
lista = mi_lista_csv(lista)
```

```python
#Ya para crear plos mapas 
def mi_mapa(archivo_mapa,ruta='./'):
    datos_mapa = np.genfromtxt(ruta+archivo_mapa, delimiter=',', dtype=float)
    t, lat_long = datos_mapa[1:,0], datos_mapa[1:,1:3]  #Juntamos lat_long pq pliyline mo quiere màs datos
    
    mapa = folium.Map(location=lat_long[0],zoom_start=14) #Genera el mapa al rededor de este punto
    folium.Marker(location=[lat_long[0,0],lat_long[0,1]], popup="Start", icon=folium.Icon(color='green')).add_to(mapa) #Marcador inicial
    folium.Marker(location=[lat_long[-1,0],lat_long[-1,1]], popup="End", icon=folium.Icon(color='orange')).add_to(mapa) #Marcador final
    trayectoria = folium.PolyLine(locations=lat_long).add_to(mapa)
    
    mapa.save(ruta+archivo_mapa[:-4]+'.html')
```

```python
ruta_csv='./ArchivosCSV/'
lista = os.listdir(ruta_csv)
lista = mi_lista_csv(lista)

for mapita in lista:
    mi_mapa(mapita,ruta=ruta_csv)
    mapa_html=ruta_csv+mapita[:-4]+'.html'
    print(mapa_html)
    marco = IFrame(src=mapa_html,width=250,height=259)
    display(marco)#Lo necesito para que me nuestre el IFrame
```

```python
#Y si ahora queremos calcular distancias, velocidades y asì
def dist_mapa(archivo_mapa,ruta='./'):
    datos_mapa = np.genfromtxt(ruta+archivo_mapa, delimiter=',', dtype=float)
    t, lat, long = datos_mapa[1:,0], datos_mapa[1:,1], datos_mapa[1:,2]
    lat, long = np.deg2rad(lat), np.deg2rad(long)
    altura = datos_mapa[1:,3]
    
    d_acumulada = np.zeros_like(t)
    RT = 6731e3
    x = (RT + altura)*np.cos(lat)*np.sin(long)
    y = (RT + altura)*np.cos(lat)*np.cos(long)
    z = (RT + altura)*np.sin(lat)

    for i in range(len(t)-1):
        d_acumulada[i+1] = (d_acumulada[i]+np.sqrt((x[i+1]-x[i])**2 + (y[i+1]-y[i])**2 + (z[i+1]-z[i])**2))

    return d_acumulada
```

```python
for cada_mapa in lista:
    d = dist_mapa(ruta_csv+cada_mapa)
    print('archivo: ', cada_mapa)
    print('dìstancia=',d[-1]/100,'km')
```
