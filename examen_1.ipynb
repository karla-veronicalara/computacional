1. Escriba un código que sume los números múltiplos de 7 y 13 que hay entre 0 y 100,000, es decir, que sume 0 + 7 + 13 + 14 + 26 + 21 + 39 + 28 + 52 + 35 + ...

*(2 puntos)*

````python
import numpy as np
import matplotlib.pyplot as plt
````

````python
m7 = np.arange(0,100000,7)     #Arreglo que contiene los múltiplos de 7
m13 = np.arange(0,100000,13)   #Arreglo que contiene los múltiplos de 13
a = 0                          #Variable donde almacenaremos el valor de la suma

#Notemos que m7 y m13 no poseen necesariamente la misma longitud, por lo que haremos un ciclo por separado para cada arreglo
for i in range(len(m7)):
    a += m7[i]
    
for j in range(len(m13)):
    a += m13[j]

print('El resultado de la suma es ', a)
````

2. Haga la gráfica de la función $f(x)=x^2 - \frac{x^4}{2} + \frac{x^6}{3} - \frac{x^8}{4} + \frac{x^{10}}{5} $ y de la función $\ln (1+x^2)$ en la misma gráfica. Como se comparan las dos funciones anteriores en el intervalo $|x|<1$. Explique las similitudes y diferencias.

*(2 puntos)*

````python
#Gráfica de las funciones
x = np.linspace(-10,10,200)
f = x**2 - (x**4)/2 + (x**6)/3 - (x**8)/4 + (x**10)/5
g = np.log(1 + x**2)

plt.plot(x,f, label=r'$\ln(1+x^2)$')
plt.plot(x,f,'--', label='f(x)')
plt.title('Gráfica funciones')

plt.legend(), plt.grid()
plt.show()
````

````python
#En el intervalo |x|<1
plt.plot(x,f, label=r'$\ln(1+x^2)$')
plt.plot(x,f,'--', label='f(x)')
plt.title(r'Gráfica funciones en el intervalo $|x|<1$')
plt.xlim(-1,1), plt.ylim(0,1)

plt.legend(), plt.grid()
plt.show()

#Notemos que ambas gráficas son prácticamente iguales, de hecho si calculamos el error:
error = abs(f-g)
plt.plot(x,error, label=r'$\ln(1+x^2)$')
plt.title('Comparación')
plt.xlim(-1,1), plt.ylim(0,0.1)
plt.grid()
plt.show()
````
